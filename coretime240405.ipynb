```python
# 각 행의 쉼표 개수를 저장할 리스트 초기화
comma_counts = []

# 데이터 시리즈에서 각 행에 대해 반복
for row in works:
    # NaN인 경우 무시하고 넘어감
    if pd.isna(row):
        comma_counts.append(0)
    else:
        # 쉼표의 개수를 세고 리스트에 추가
        comma_counts.append(row.count(','))

print(comma_counts)
```

```python
def format_time(time_str):
    if pd.isna(time_str):
        return time_str
    else:
        # 쉼표를 기준으로 분리한 후 각 요소를 [시작시간, 종료시간] 형태로 반환
        times = time_str.split(',')
        formatted_times = []
        for t in times:
            start, end = t.split('~')
            formatted_times.append([start, end])
        return formatted_times

# 각 요소에 함수를 적용하여 변경된 결과를 새로운 시리즈에 저장
formatted_data = works.apply(format_time)

print(formatted_data)
```
